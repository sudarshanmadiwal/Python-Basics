### Regular expressions
- a set of instructions used to search for a particular pattern in a txt/file
- it is a set of combination of special chars with specific number
- common for most of the languages
- though the meta-chars are common across the languages the functions to use them are different for each language
- python uses the library re for regular expressions

### search() function in this module, which is used to find particular substring within a string.Consider the Following example

```python
import re

hand = open('regular.txt')

for line in hand:
    line = line.rstrip()
    if re.search('how',line):
        print(line)
        
hand.close()
```

Every line is searched with charecters 'how' in the every line if 'how' is presented in the line, then line is printed

```python
#search the lines begining with 'how' then any two chars and end with m 
fhand = open('regular.txt')
for line in fhand:
    line = line.strip()
    if re.search('^how',line):# we make use'^' symbol which indicates beginning of the line
        print(line)
fhand.close()    
```

```python
# starting with h followed by one or more chars and then u somewherein the line

fhand = open('regular.txt')
for line in fhand:
    line = line.strip()
    if re.search('^h.+u',line):
        print(line)
fhand.close()
```

- ^ (caret) Matches begiinning of the line
- $ Matches end of the line
- . (dot) Macthes any single charecter except newline. using option m, then newline also can be matched
- [...] Matches any single chareter in brackets
- [^...] Matches any single chareter in NOT brackets
- re* Matches 0 or More occurences of preceding expression
- re+ Matches 1 or more occurences of preceding expression.
- re{n} Matches exactly n number of occurences of preceding expression
- re?
- re{n,} matches n or more occurences of preceding expression
- re{n,m} matches at least n and at most m occurrences of preceding expression
- a|b Matches either a or b
- (re) Groups regular expressions and remebmers matches text
- \d Matches digits.equivalent to [0-9]
- \D MAtches non digits.
- \w matches word charecters
- \W matches non word charecters
- \s matches whitespace
- \S amatches non whitespace
- \A Matches beginning of string.
- \Z Matches end of string. if a newline exists, it matches just before newline.
- \z Matches end of string


#### This module exports the following functions:
- match     -Match a regular expression pattern to the beginning of a string.
- fullmatch -Match a regular expression pattern to all of a string.
- search    -Search a string for the presence of a pattern.
- sub       -Substitute occurrences of a pattern found in a string.
- subn      -Same as sub, but also return the number of substitutions made.
- split     -Split a string by the occurrences of a pattern.
- findall   -Find all occurrences of a pattern in a string.
- finditer  -Return an iterator yielding a Match object for each match.
- compile   -Compile a pattern into a Pattern object.
- purge     -Clear the regular expression cache.
- escape    -Backslash all non-alphanumerics in a string.

```python
# extrqact all the lines ending with any digit
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.strip()
    if re.search('[0-9]$',line):
        print(line)
fhand.close()
```

```python
# extract the lines starting with anything othwer than lwer case alphabets and digits
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    if re.search('[^a-z0-9]',line):
        print(line)
fhand.close()
```

```python
#demo of fidall() method
#extract email Id from a given string

s = 'a message from sudarshanmadiwal@gmail.com to fsdfn@educom.in  at @4pm'
lst = re.findall('\S+@\S+',s)# \S matches non white spaces
print(lst) 
#extract only domain
for domain in lst:
    x=domain.find('@')
    print(x)
    print(domain[x+1:])
    
```

```python
fhand = open('mbox-short.txt')
ls =[]
for line in fhand:
    line = line.lstrip()
#     lst = re.findall('[a-zA-Z0-9._]\S+@\S+[a-zA-Z]',line)
    lst = re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-Z][.][a-zA-Z]+',line)
    if len(lst)>0:
        ls.extend(lst)
print(ls)        
fhand.close()
```
